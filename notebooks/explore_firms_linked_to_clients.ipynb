```python
import pandas as pd

# 1. Load the processed Overture CSV
df = pd.read_csv('../data/processed/overture_life_diligence_data.csv')

# 2. Convert dates to actual Pandas Datetime objects
df['Filing_Date'] = pd.to_datetime(df['Filing_Date'])

# 3. Create a clean summary table showing when each Law Firm was active
timeline = df.groupby('Law_Firm').agg(
    First_Patent_Filed=('Filing_Date', 'min'),
    Last_Patent_Filed=('Filing_Date', 'max'),
    Total_Cases=('App_Number', 'count')
).sort_values('First_Patent_Filed')

# 4. Make it look pretty
print("📅 OVERTURE LIFE: LAW FIRM TIMELINE")
print("-" * 50)
display(timeline)
```
